# Agora vamos olhar para os DOADORES (atualizado em 22.09.2022)

```python
# agora vamos olhar para os DOADORES de campanha. 
# A fonte é o arquivo "receitas_candidatos_2022_BRASIL", também do TSE, que pode ser baixado aqui: https://dadosabertos.tse.jus.br/dataset/dadosabertos-tse-jus-br-dataset-prestacao-de-contas-eleitorais-2022)
# no nosso script, vamos baixar os dados mais atuais diretamente do link, sem necessidade de baixar o arquivo no computador

from pandas.io.common import codecs
import requests, zipfile, io
import pandas as pd

url = 'https://cdn.tse.jus.br/estatistica/sead/odsele/prestacao_contas/prestacao_de_contas_eleitorais_candidatos_2022.zip'
filename = 'receitas_candidatos_2022_BRASIL.csv'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}

r = requests.get(url, headers=headers)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()

doadores = pd.read_csv(filename, sep=';', encoding='latin1', converters={'NR_CPF_CANDIDATO': str, 'NR_CPF_CNPJ_DOADOR': str}, decimal=',')

doadores.head(5)
```

```python
# criar coluna nova que fala quem é o doadores e quem é o candidato que recebeu a doação, para depois sabermos o total doado a cada candidato por cada doador
doadores['cpf_doador__cpf_candidato'] = doadores.NR_CPF_CNPJ_DOADOR + '_' + doadores.NR_CPF_CANDIDATO
doadores['nome_doador__nome_candidato'] = doadores.NM_DOADOR + '_' + doadores.NM_CANDIDATO
```

```python
# criar tabela nova que mostra o total que foi doado por aquele doador a aquele candidato até o dia 15.09.2022
total_doado = doadores.groupby(['cpf_doador__cpf_candidato', 'nome_doador__nome_candidato']).VR_RECEITA.sum().round(2).reset_index().sort_values('VR_RECEITA', ascending=False)

total_doado['cpf_cnpj_doador'] = total_doado.cpf_doador__cpf_candidato.apply(lambda x: x.split('_')[0])
total_doado['cpf_candidato'] = total_doado.cpf_doador__cpf_candidato.apply(lambda x: x.split('_')[1])
total_doado['nome_doador'] = total_doado.nome_doador__nome_candidato.apply(lambda x: x.split('_')[0])
total_doado['nome_candidato'] = total_doado.nome_doador__nome_candidato.apply(lambda x: x.split('_')[1])

total_doado = total_doado.rename(columns={'VR_RECEITA': 'total_doado_pelo_doador_a_este_candidato'})

total_doado = total_doado[['cpf_cnpj_doador',	'cpf_candidato', 'nome_doador',	'nome_candidato', 'total_doado_pelo_doador_a_este_candidato']]
total_doado.head(10)
```

```python
# remover da lista os doadores que nao apresentaram CPF/nome
doadores = doadores[doadores.NR_CPF_CNPJ_DOADOR != '-1']
```

```python
#vamos unificar estas novas colunas à tabela geral de doações
# tabela final de doadores: "doadores_com_totais"

doadores_com_totais = doadores.merge(total_doado, left_on=['NR_CPF_CNPJ_DOADOR', 'NR_CPF_CANDIDATO'], right_on= ['cpf_cnpj_doador', 'cpf_candidato'])

doadores_com_totais.sort_values('total_doado_pelo_doador_a_este_candidato', ascending=False).head(3)
```

## Vamos descobrir quais doadores tiveram multas ambientais e quantas são

```python
multas_doadores = naonulos.merge(doadores, left_on='cpf_ou_cnpj', right_on='NR_CPF_CNPJ_DOADOR')

multas_doadores.count()
```

```python
multas_doadores.groupby(['nome_autuado', 'cpf_ou_cnpj']).valor_multa.sum().reset_index().sort_values('valor_multa', ascending=False).head(20)
```
